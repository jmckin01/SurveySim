```python
import numpy as np
from scipy.optimize import curve_fit
from pylab import *
%matplotlib inline 

def func(x,m,b,n):
    return (1.0+np.exp((m-x)/abs(b)))**(-n)

def fitCompCurve(flux,completeness,label=''):
    popt, pcov = curve_fit(func, flux, completeness, np.array([9.0,5.0,1.48]),maxfev=10000)
    scatter(flux,completeness)
    x=np.arange(0.1,320.0,0.1)
    plot(x,func(x,popt[0],popt[1],popt[2]),label=label)
    return popt
```

```python
s = np.array([5.0, 10.0, 20.0, 40.0, 80.0, 160.0])
c250 = np.array([0.19, 0.39, 0.80, 0.99, 1.0, 1.0])
c350 = np.array([0.33, 0.61, 0.93, 1.0, 1.0, 1.0])
c500 = np.array([0.73, 0.91, 0.99, 1.0, 1.0, 1.0])

figure(figsize=(6,5))
p250=fitCompCurve(s,c250,label='250')
print p250
p350=fitCompCurve(s,c350,label='350')
print p350
p500=fitCompCurve(s,c500,label='500')
print p500

xscale('log')
xlim(1.0,320.0)
legend(loc='upper left',title='Band')
xlabel('Flux (mJy)')
ylabel('Completeness')
title('COSMOS')
savefig('cosmos.png')
```
