```python
import numpy
from pylab import *
```

```python
model,z,type,filter,flux = numpy.loadtxt("fluxes.txt",unpack=True)
```

```python
models=numpy.unique(model)
print(models)
for i in range(0,len(models)):
    inds = where((model == models[i]) & (type == 0) & (filter == 0))
    print(inds)
    zplot=z[inds]
    fplot=flux[inds]
    plot(z[inds],flux[inds],label=str(models[i]))
    yscale('log')
    legend()
    show()
```
