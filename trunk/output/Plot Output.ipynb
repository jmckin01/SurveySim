```python
def imgcontour (data,levels,extent):
    cset = contour(data, levels, origin="lower",colors="black",extent=extent)
    clabel(cset, inline=1, fontsize=10, fmt="%2.0i")
    for c in cset.collections:
        c.set_linestyle("solid")
```

```python
extent = [hdus[0].header['H_MIN_Y'], hdus[0].header['H_MAX_Y'], hdus[0].header['H_MIN_X'], hdus[0].header['H_MAX_X']]
print(hdus[1].data.max())
levels=(30,20,10,2)
reslevels=(10,1)

figure(1,figsize=(12,4))
for i in range(0,3):
    subplot(1,3,i+1)
    print(hdus[i].data.sum())
    if(i==0):
        title('Residual')
        ylabel('Axis 2')
        imgcontour(hdus[i].data,reslevels,extent)
    else:
        if(i==1):
            title('Model')
        else:
            title('Observation')
        imgcontour(hdus[i].data,levels,extent)
    xlabel('Axis 1')
    xlim(0.5,2.5)
    ylim(-0.75,0.75)
show()
```

```python
dists=hdus[3].data
#hdus[3].data.names
chainchi=hdus[7].data['chisq']
chimed=median(chainchi)
gpts=numpy.where(chainchi < chimed)

s1=dists['s1']
od1=dists['obs_dnds1']
md1=dists['mod_dnds1']

d1chain=hdus[7].data['dnds1'][gpts]
mins1=numpy.zeros(len(s1))
maxs1=numpy.zeros(len(s1))
for i in range(0,len(d1chain[0])):
    mins1[i]=min(d1chain[:,i])
    maxs1[i]=max(d1chain[:,i])

pts1=where(md1 > 1.0)
plot(s1[pts1],od1[pts1],'o',label="Observation")
plot(s1[pts1],md1[pts1],'o',label="Model")
errorbar(s1[pts1],md1[pts1],yerr=[mins1[pts1],maxs1[pts1]],linestyle="None")
title("Counts Band 1")
xlabel('Flux [mJy]')
ylabel('DnDs*Jy^(-1.5)')
xscale('log')
yscale('log')
legend(loc='lower right')
ylim(1e0,1e5)
show()

s2=dists['s2']
od2=dists['obs_dnds2']
md2=dists['mod_dnds2']

d2chain=hdus[7].data['dnds2'][gpts]
std2=numpy.zeros(len(s2))
for i in range(0,len(d2chain[0])):
    std2[i]=std(d2chain[:,i])

pts2=where(md2 > 1.0)
plot(s2[pts2],od2[pts2],'o',label="Observation")
plot(s2[pts2],md2[pts2],'o',label="Model")
errorbar(s2[pts2],md2[pts2],yerr=std2[pts2],linestyle="None")
title("Counts Band 2")
xlabel('Flux [mJy]')
ylabel('DnDs*Jy^(-1.5)')
xscale('log')
yscale('log')
legend(loc='lower right')
ylim(1e0,1e5)
show()

s3=dists['s3']
od3=dists['obs_dnds3']
md3=dists['mod_dnds3']

d3chain=hdus[7].data['dnds3'][gpts]
std3=numpy.zeros(len(s3))
for i in range(0,len(d3chain[0])):
    std3[i]=std(d3chain[:,i])

pts3=where(md3 > 1.0)
plot(s3[pts3],od3[pts3],'o',label="Observation")
plot(s3[pts3],md3[pts3],'o',label="Model")
errorbar(s3[pts3],md3[pts3],yerr=std3[pts3],linestyle="None")
title("Counts Band 3")
xlabel('Flux [mJy]')
ylabel('DnDs*Jy^(-1.5)')
xscale('log')
yscale('log')
legend(loc='lower right')
ylim(1e0,1e5)
show()
```

```python
def bins (dat): return int(ceil(pow(len(dat),0.333)))
def binSize (dat): return 2.4*std(dat)/pow(len(dat),0.333)
def nbins (dat): return int(ceil((max(dat)-min(dat))/binSize(dat)))
def nbins2 (dat): return int(ceil((max(dat)-min(dat))/binSize(dat)/2))
```

```python
print(len((dists['Lum'])))
print(nbins(dists['z']))
print(binSize(dists['z']))
```

```python
def contour2d (x,y):
    H, xedges, yedges = histogram2d(y, x, bins=(nbins2(y), nbins2(x)),normed=True)
    extent = [yedges[0], yedges[-1], xedges[0], xedges[-1]]
    levels = (0.5,0.4,0.3,0.2,0.15,0.1,0.075,0.06,0.05,0.035,0.02)
    cset = contour(H, levels, origin="lower",linewidths=(1.9, 1.6, 1.5, 1.4),colors="black",extent=extent)
    clabel(cset, inline=1, fontsize=10, fmt="%0.2f")
    for c in cset.collections:
        c.set_linestyle("solid")
```

```python
params=hdus[4].data
cols=params.names
chain_num=0
param_num=0
pnames=[]
for col in cols:
    if "0" in col:
        if not (("CHISQ" in col) | ("ACPT" in col)):
            ptemp=col.split('0')[0]
            pnames.append(ptemp)
    if "CHISQ" in col:
        chain_num+=1

pchains=[]
for i in range(0,len(pnames)):
    p=pnames[i]
    pdat=[]
    for i in range(0,chain_num):
        pdat.extend(params[p+str(i)])
    pchains.append(pdat)
    
figure(figsize=(8,8))
for i in range(0,len(pnames)):
    for j in range(0,len(pnames)):
        if(i >= j):
            subplot(len(pnames),len(pnames),i*len(pnames)+j+1)
        if(i == j):
            hist(pchains[i],bins=nbins(pchains[i]),normed=True,histtype='step',color="black")
            if(j == 0):
                ylabel("Relative Probability")
        elif(i > j):
            contour2d(pchains[j],pchains[i])
            if(j == 0):
                ylabel(pnames[i])
        if(i == (len(pnames)-1)):
            xlabel(pnames[j])
show()
```
