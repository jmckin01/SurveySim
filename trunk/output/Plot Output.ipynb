```python
%matplotlib inline
import pyfits
import numpy
from pylab import *
```

```python
hdus=pyfits.open("output.fits")
```

```python
hdus.info()
```

```python
matshow(hdus[1].data)
show()
matshow(hdus[2].data)
show()
matshow(hdus[0].data)
show()
```

```python
dists=hdus[3].data
#hdus[3].data.names
chainchi=hdus[7].data['chisq']
chimed=median(chainchi)
gpts=numpy.where(chainchi < chimed)

s1=dists['s1']
od1=dists['obs_dnds1']
md1=dists['mod_dnds1']

d1chain=hdus[7].data['dnds1'][gpts]
std1=numpy.zeros(len(s1))
for i in range(0,len(d1chain[0])):
    std1[i]=std(d1chain[:,i])

pts1=where(od1 > 1.0)
plot(s1[pts1],od1[pts1],'o',label="Observation")
plot(s1[pts1],md1[pts1],'o',label="Model")
errorbar(s1[pts1],md1[pts1],yerr=std1[pts1],linestyle="None")
title("Counts Band 1")
xlabel('Flux [mJy]')
ylabel('DnDs*Jy^(-1.5)')
xscale('log')
yscale('log')
legend(loc='lower right')
ylim(1e0,1e5)
show()

s2=dists['s2']
od2=dists['obs_dnds2']
md2=dists['mod_dnds2']

d2chain=hdus[7].data['dnds2'][gpts]
std2=numpy.zeros(len(s2))
for i in range(0,len(d2chain[0])):
    std2[i]=std(d2chain[:,i])

pts2=where(od2 > 1.0)
plot(s2[pts2],od2[pts2],'o',label="Observation")
plot(s2[pts2],md2[pts2],'o',label="Model")
errorbar(s2[pts2],md2[pts2],yerr=std2[pts2],linestyle="None")
title("Counts Band 2")
xlabel('Flux [mJy]')
ylabel('DnDs*Jy^(-1.5)')
xscale('log')
yscale('log')
legend(loc='lower right')
ylim(1e0,1e5)
show()

s3=dists['s3']
od3=dists['obs_dnds3']
md3=dists['mod_dnds3']

d3chain=hdus[7].data['dnds3'][gpts]
std3=numpy.zeros(len(s3))
for i in range(0,len(d3chain[0])):
    std3[i]=std(d3chain[:,i])

pts3=where(od3 > 1.0)
plot(s3[pts3],od3[pts3],'o',label="Observation")
plot(s3[pts3],md3[pts3],'o',label="Model")
errorbar(s3[pts3],md3[pts3],yerr=std3[pts3],linestyle="None")
title("Counts Band 3")
xlabel('Flux [mJy]')
ylabel('DnDs*Jy^(-1.5)')
xscale('log')
yscale('log')
legend(loc='lower right')
ylim(1e0,1e5)
show()
```
