```python
%matplotlib inline
import pyfits
from pylab import *
```

```python
hdus = pyfits.open("my_templates.fits")
templates=hdus[0].data
x=(templates[0,0,0])
for i in range(1,len(templates[0,0])):
    y=templates[0,0,i]
    plot(x,y)

xscale('log')
yscale('log')
xlim(1e-1,1e6)
show()

hdus = pyfits.open("sf_templates.fits")
templates=hdus[0].data
x=(templates[0])
for i in range(1,len(templates)):
    y=templates[i]
    plot(x,y)

xscale('log')
yscale('log')
xlim(1e-1,1e6)
show()
```
